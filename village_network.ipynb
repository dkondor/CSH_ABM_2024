```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vec1 = pd.read_csv('demog_vectors.csv')
```

```python
df = vec1
plt.figure(figsize=(10, 6))
plt.plot(df['pstar'], label='Survival Probability', color='blue', linestyle='-', marker='o')
plt.plot(df['mstar'], label='Fertility Probability', color='green', linestyle='--', marker='x')

# Adding titles and labels
plt.title('Survival and Fertility Probabilities', fontsize = 20)
plt.xlabel('Age Class', fontsize = 20)  
plt.ylabel('Probability', fontsize = 20)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.legend(fontsize = 15)

plt.grid(True)
plt.show()
```

```python
import json
import ast
def read_list_of_dicts(file_name):
    with open(file_name, 'r') as file:
        content = file.read()
        data = ast.literal_eval(content)
        
    return data

file_name = 'networks2000.txt'  
data = read_list_of_dicts(file_name)
```

```python
import csv
year = 1999
households = data[year]
with open(f'nodes{year}.csv', mode='w', newline='') as nodes_file:
    writer = csv.writer(nodes_file)
    writer.writerow(['ID', 'Label', 'num_member'])
    for household, data in households.items():
        writer.writerow([household, f'Household {household}', data['num_member']])

with open(f'edges{year}.csv', mode='w', newline='') as edges_file:
    writer = csv.writer(edges_file)
    writer.writerow(['Source', 'Target', 'Weight'])
    for household, data in households.items():
        for neighbor, connectivity in data['connectivity'].items():
            writer.writerow([household, neighbor, round(connectivity, 3)]) 
```
